# Export e Visualização

Este notebook mostra como exportar artefatos (GeoJSON, HTML) a partir de um run existente e abrir o mapa (salvo em `route_map.html`).

## 1) Escolher run e executar export

```python
# Substitua pelo run desejado. Ex.: experiments/run_1770494072
run_dir = sorted(__import__('glob').glob('experiments/run_*'))[-1]
print('Using run:', run_dir)
!PYTHONPATH=. python3 scripts/export_artifacts.py {run_dir}
!ls -la {run_dir} | sed -n '1,200p'
```

## 2) Visualizar o GeoJSON (conteúdo resumido)

```python
import json
with open(run_dir + '/routes.geojson','r',encoding='utf-8') as f:
    gj = json.load(f)
print('Features:', len(gj.get('features', [])))
print('Sample feature props:')
print(gj['features'][0]['properties'])
```

## 3) Abrir mapa HTML (clique no link do notebook ou abra localmente)

```python
from IPython.display import HTML, display
html_path = run_dir + '/route_map.html'
print('Mapa salvo em:', html_path)
# Exibir iframe (pode não renderizar em alguns ambientes)
display(HTML(f"<iframe src='{html_path}' width='100%' height='600px'></iframe>"))
```
