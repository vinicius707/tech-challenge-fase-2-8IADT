# Compare Results — Aggregate Experiments

Este notebook carrega `experiments/batch_runs/aggregate_results.csv` e gera gráficos comparativos para analisar o desempenho (best_fitness) entre configurações.

```python
import pandas as pd
import matplotlib.pyplot as plt

PATH = 'experiments/batch_runs/aggregate_results.csv'

df = pd.read_csv(PATH)

df['name_short'] = df['name']
print('Aggregate results:')
df_display = df[['name','population','generations','mutation_rate','best_fitness']]
display(df_display)

# Plot best_fitness by experiment name
plt.figure(figsize=(8,4))
plt.bar(df['name'], df['best_fitness'], color='tab:blue')
plt.ylabel('best_fitness (lower is better)')
plt.title('Comparison of Best Fitness across experiments')
plt.grid(axis='y', alpha=0.3)
plt.show()
```
